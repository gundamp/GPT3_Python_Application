```python
model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = pipeline (

    "text-generation", #Task name
    model=model,
    tokenizer=tokenizer,
    torch_dtype = torch.bfloat16, #data type
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id

)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature': 0.6})
```

```python
template = " ""You are an intelligent chatbot. Help the following question with brilliant answers. Question: {question} Answer: "" "

prompt = PromptTemplate(template=template, input_variables=["question"])

llm_chain = LLMChain(prompt=prompt, llm=llm)

question = " Can you help me understand a bit about what Dai-ichi Life does? "

print(llm_chain.run(question))
```

```python
template = "What is a good name for a company that makes {product}?"

prompt = PromptTemplate(
    input_variables=["product"],
    template=template,
)

#prompt.format(product="colorful socks")

llm_chain = LLMChain(prompt=prompt, llm=llm)

print(llm_chain.run("colorful socks"))
```

```python
from langchain.chains import LLMChain, SimpleSequentialChain

# Define the first chain as in the previous code example
# ...

# Create a second chain with a prompt template and an LLM
second_prompt = PromptTemplate(
    input_variables=["company_name"],
    template="Write a catchphrase for the following company: {company_name}",
)

chain_two = LLMChain(llm=llm, prompt=second_prompt)

# Combine the first and the second chain
overall_chain = SimpleSequentialChain(chains=[llm_chain, chain_two], verbose=True)

# Run the chain specifying only the input variable for the first chain.
catchphrase = overall_chain.run("colorful socks")
```

```python
from langchain import PromptTemplate, FewShotPromptTemplate

examples = [
    {"word": "happy", "antonym": "sad"},
    {"word": "tall", "antonym": "short"},
]

example_template = """
Word: {word}
Antonym: {antonym}\n
"""

example_prompt = PromptTemplate(
    input_variables=["word", "antonym"],
    template=example_template,
)

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Give the antonym of every input",
    suffix="Word: {input}\nAntonym:",
    input_variables=["input"],
    example_separator="\n",
)

few_shot_prompt.format(input="big")

llm_chain = LLMChain(prompt=few_shot_prompt, llm=llm)

print(llm_chain.run("skinny"))
```
